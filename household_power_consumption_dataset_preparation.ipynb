```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values = []

with open('data/household_power_consumption.txt') as fp:
    fp.readline() # skip header
    
    line = fp.readline()
    while line:
        splited = line[:-1].split(";")
        try:
            a = float(splited[2])
            b = float(splited[3])
            c = float(splited[4])
            d = float(splited[5])
            values.append(( a, b, c, d ))
        except Exception:
            print(splited)
            
        line = fp.readline()
```

```python
d = list(map(lambda x: x[3], values))
```

```python
ts = []
for i in range(0, len(d)-2048, 256):
    ts.append(d[i:i+2048])
```
